```python
def plot(x, title='Plot of DT signal x[n]'):
  """
  x: DT signal x[n], 1D array
  """
  n = np.arange(x.shape[0])
  plt.xlabel('n');
  plt.ylabel('x[n]');
  plt.title(title);
  plt.stem(n, x, use_line_collection=True);
  plt.show()
```

#### Reconstruct the above signal.

Use the following link to learn how to use the np.cos function:

https://numpy.org/doc/stable/reference/generated/numpy.cos.html

```python
# your code here
A =  3
phi =  pi
omega =  pi/16

n =  100
x =  np.array([A * np.cos(omega*i + phi) for i in range(n)])
plot(x)
```
